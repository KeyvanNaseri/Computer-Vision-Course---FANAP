### Exercise 15

```python
class Dog:

    def __init__(self, Name, Breed, Size, Age, Color):
        self.Name = Name
        self.Breed = Breed
        self.Size = Size
        self.Age = Age
        self.Color = Color
        
    def Eat(self):
        return 'Eat your food {}'.format(self.Name)
    
    def Sleep(self):
        return 'Sleep {}'.format(self.Name)
   
    def Sit(self):
        return '{}, Sit'.format(self.Name)
    
    def Run(self):
        return 'Come on, Run {}'.format(self.Name)
    
    def __str__(self):
        return f'Breed of the dog is {self.Breed}, It is {self.Size} and {self.Color}, the dog is {self.Age} years old'
```

```python
dog1 = Dog('Bella', 'Beaglier', 'Small', 2, 'red')
dog1.__str__()
```
