```python
#1
dataset = []
all_file_names = os.listdir('dataset/images')
for filename in all_file_names:
    image = cv2.imread('dataset/images/' + filename,-1)
    dataset.append(image)
```

```python
#2
def image_type(image):
    
    unique_pixel_values = set(image.flatten())
    
    if len(image.shape) == 3 and image.shape[2]==3:
        return "Color"
        
    elif len(image.shape) == 2:
        
        if len(unique_pixel_values) < 20:
            return "Binary"
        else:
            return "Grayscale"
    else:
        return "Unknown"

for i in range(len(dataset)):
    print('The type of the image with the name {} is {}'.format(all_file_names[i],image_type(dataset[i])))
```

```python
#3
c2g = []
for i,image in enumerate(dataset):
    if image_type(image) == 'Color':
        gray_im = cv2.cvtColor(dataset[i], cv2.COLOR_BGR2GRAY)
        op_gray_im = 255 - gray_im
        c2g.append(op_gray_im)
        plt.figure()
        plt.imshow(op_gray_im,'gray')
```

```python
#4
g2b = []
for i,image in enumerate(dataset):
    if image_type(image) == 'Grayscale':
        bin_im = cv2.threshold(dataset[i], 128, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]
        op_bin_im = 255 - bin_im
        g2b.append(op_bin_im)
        plt.figure()
        plt.imshow(op_bin_im,'gray')
```

```python
#5
op_b = []
for i,image in enumerate(dataset):
    if image_type(image) == 'Binary':
        op_bin = 255 - image
        op_b.append(op_bin)
        plt.figure()
        plt.imshow(op_bin,'gray')
```

```python
y = 512*10*10**(-6)   # Height of the image
Y = 0.75/0.9          # Height of the object to ensure that its height fills 90 percent of the FOV
o = 4                 # Distance between the object and the lens
i = o*y/Y
f = np.reciprocal(i**(-1) + o**(-1))
print('The focal length is equal to {r:1.6f}'.format(r=f))
```

```python
power_1 = 40  # diopters
power_2 = 10  # diopters

# Calculate total power
power_total = power_1 + power_2

# Calculate total focal length
focal_length_total = 1 / power_total

print("The total focal length of the lens system is: {f:.2f} meters".format(f = focal_length_total))
```
