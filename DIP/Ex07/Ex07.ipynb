```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('sample.jpg', -1)
```

```python
clahe =cv2.createCLAHE(clipLimit=95,tileGridSize=(2,2))
clahe_img = clahe.apply(img)
```

```python
plt.figure(figsize=[12,8])
plt.subplot(221);plt.imshow(img, cmap='gray');                plt.title("Original image");
plt.subplot(222);plt.hist(img.ravel(), 256);                  plt.title("Original image histogram");

plt.subplot(223);plt.imshow(clahe_img, cmap='gray');             plt.title("CLAHE image");
plt.subplot(224);plt.hist(clahe_img.ravel(), 256);               plt.title("CLAHE image histogram");
```
