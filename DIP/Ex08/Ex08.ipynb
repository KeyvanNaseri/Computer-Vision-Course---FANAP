### PNG Image

```python
image_png = cv2.imread('images for ex08/input.png', -1)

clahe =cv2.createCLAHE(clipLimit=15,tileGridSize=(8,8))
channels = cv2.split(image_png); eq_channels = []

for ch in channels:
    ch = cv2.medianBlur(ch, ksize=5)
    ch= clahe.apply(ch)
    ch = cv2.GaussianBlur(ch,(3,3), 0)
    eq_channels.append(ch)

enhanced_image_png = cv2.merge(eq_channels)
```

```python
plt.figure(figsize=[10,4]);
plt.subplot(121); plt.imshow(image_png); plt.subplot(122); plt.imshow(enhanced_image_png); plt.show()

color = ('b', 'g', 'r')
plt.figure(figsize=[4,3]);plt.title('Histogram')
for i, col in enumerate(color):
    histogram = cv2.calcHist([enhanced_image_png], [i], None, [256], [0, 256])
    plt.plot(histogram, color = col)
```

### JPG Image

```python
def contrast_clipping(img,clip_ratio=0.01):
    sorted_pixels = np.sort(img.ravel())
    num_delete = int(len(sorted_pixels)*clip_ratio)
    min_intensity = sorted_pixels[num_delete]
    max_intensity = sorted_pixels[len(sorted_pixels)-num_delete]
    cc_img = img.copy()
    cc_img[cc_img>max_intensity] = 255; cc_img[cc_img<min_intensity] = 0
    cc_img = np.uint8(((cc_img-min_intensity)/(max_intensity-min_intensity))*255)
    return cc_img
```

```python
image_jpg = cv2.imread('images for ex08/input.jpg', -1)

channels2 = cv2.split(image_jpg); eq_channels2 = []
clahe2 =cv2.createCLAHE(clipLimit=1,tileGridSize=(8,8))


for ch in channels2:
    ch = cv2.medianBlur(ch, ksize=11)
    ch = contrast_clipping(ch,clip_ratio=0.005)
    ch = cv2.GaussianBlur(ch,(5,5), 0)
    eq_channels2.append(ch)

enhanced_image_jpg = cv2.merge(eq_channels2)
```

```python
plt.figure(figsize=[10,4]);
plt.subplot(121); plt.imshow(image_jpg[:,:,::-1]); plt.subplot(122); plt.imshow(enhanced_image_jpg[:,:,::-1]); plt.show()

color = ('b', 'g', 'r')
plt.figure(figsize=[4,3]);plt.title('Histogram')
for i, col in enumerate(color):
    histogram = cv2.calcHist([enhanced_image_jpg], [i], None, [256], [0, 256])
    plt.plot(histogram, color = col)
```

### JPEG is a lossy compression format, which means it achieves high compression ratios by discarding some image information. This loss of information can result in a reduction in image quality, especially at higher compression levels. As a result of that, the histogram may show fewer distinct intensity and more smoothness levels due to quantization effects introduced during compression. So, the image shows less contrast and less quality.
